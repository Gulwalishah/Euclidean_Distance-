# Euclidean distance function

```python
import math  # Import the math module to use the square root function

# Define a function to calculate Euclidean distance
def euclidean_distance(x1, y1, x2, y2):
    # Calculate the differences between x and y coordinates
    x_diff = x2 - x1
    y_diff = y2 - y1

    # Square the differences
    x_diff_squared = x_diff ** 2
    y_diff_squared = y_diff ** 2

    # Sum the squared differences
    sum_of_squares = x_diff_squared + y_diff_squared

    # Take the square root of the sum to get the Euclidean distance
    distance = math.sqrt(sum_of_squares)

    return distance

# Example usage
x1, y1 = 1, 2
x2, y2 = 4, 6

distance = euclidean_distance(x1, y1, x2, y2)
print(f"The Euclidean distance between ({x1}, {y1}) and ({x2}, {y2}) is {distance:.2f}")
```

```python
import math
import matplotlib.pyplot as plt

def euclidean_distance(x1, y1, x2, y2):
    x_diff = x2 - x1
    y_diff = y2 - y1
    x_diff_squared = x_diff ** 2
    y_diff_squared = y_diff ** 2
    sum_of_squares = x_diff_squared + y_diff_squared
    distance = math.sqrt(sum_of_squares)
    return distance

# Example points
x1, y1 = 1, 2
x2, y2 = 4, 6

distance = euclidean_distance(x1, y1, x2, y2)

# Create a scatter plot of the two points
plt.scatter([x1, x2], [y1, y2], c='red', label='Points')
plt.text(x1, y1 - 0.5, f'({x1}, {y1})', fontsize=12, ha='right')
plt.text(x2, y2 + 0.5, f'({x2}, {y2})', fontsize=12, ha='right')

# Draw a line connecting the two points
plt.plot([x1, x2], [y1, y2], 'b--', label='Distance')

# Annotate the Euclidean distance
plt.text((x1 + x2) / 2, (y1 + y2) / 2, f'Distance: {distance:.2f}', fontsize=12, ha='center')

# Set axis labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Euclidean Distance Calculation')

# Add a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
